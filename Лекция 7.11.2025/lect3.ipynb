```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import pandas as pd
```

```python
data=pd.read_csv('/content/CarPrice_Assignment.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
data.drop(['car_ID','CarName'],axis = 1,inplace = True)
# Кодируем категориальные переменные
data_encoded = data.copy()
label_encoders = {}
for col in data_encoded.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    data_encoded[col] = le.fit_transform(data_encoded[col])
    label_encoders[col] = le
```

```python
data_encoded
```

```python
X = data_encoded.drop(['price'],axis=1)
Y = data_encoded['price']
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3,random_state=44)
```

Описание: Простая линейная регрессия

Особенности:

Быстро обучается.

Плохо работает хорошо с мультиколлинеарностью и выбросами.
     

```python
linearModel = LinearRegression()
linearModel.fit(X_train,Y_train)
pred = linearModel.predict(X_test)
linearError=mean_squared_error(Y_test, pred)**0.5
print(linearError)
```

Описание: Линейная регрессия с L2-регуляризацией (штраф за большие веса).

Особенности:

Устойчива к мультиколлинеарности.

Помогает избежать переобучения.

```python
ridgeModel = Ridge(alpha=200.0)
ridgeModel.fit(X_train,Y_train)
pred = ridgeModel.predict(X_test)
ridgeError=mean_squared_error(Y_test, pred)**0.5
print(ridgeError)
```

Lasso (L1-регуляризация)

Описание: Линейная регрессия с L1-регуляризацией.

Особенности:

Может обнулять коэффициенты → отбор признаков.

Полезна при большом количестве нерелевантных признаков.

```python
lassoModel=Lasso(alpha=200)
lassoModel.fit(X_train,Y_train)
pred = lassoModel.predict(X_test)
lassoError=mean_squared_error(Y_test, pred)**0.5
print(lassoError)
```

Описание: Комбинация L1 и L2 регуляризации.

Особенности:

Баланс между отбором признаков (L1) и устойчивостью (L2).

Хорошо работает при коррелированных признаках.
     

```python
ElasticNetModel= ElasticNet(alpha=0.1, l1_ratio=0.5)
ElasticNetModel.fit(X_train,Y_train)
pred = ElasticNetModel.predict(X_test)
elasticError=mean_squared_error(Y_test, pred)**0.5
print(elasticError)
```

Описание: Ансамбль деревьев решений (бэггинг).

Особенности:

Устойчива к выбросам и не требует масштабирования признаков.

Хорошо работает "из коробки".

Подвержена переобучению при большом количестве деревьев и глубоких деревьях.
     

```python
forestModel = RandomForestRegressor(n_estimators=100, max_depth=10)
forestModel.fit(X_train,Y_train)
pred = forestModel.predict(X_test)
forestError=mean_squared_error(Y_test, pred)**0.5
print(forestError)
```

KNeighborsRegressor

Описание: Регрессия на основе k ближайших соседей.

Особенности:

Непараметрическая модель.

Зависит от расстояний → требует масштабирования.

Медленное предсказание на больших данных.
     

```python
neibhorModel = KNeighborsRegressor(n_neighbors=2,weights="distance")
neibhorModel.fit(X_train,Y_train)
pred = neibhorModel.predict(X_test)
neibhorError=mean_squared_error(Y_test, pred)**0.5
print(neibhorError)
```
