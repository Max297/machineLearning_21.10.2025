Создайте консольный список покупок.У пользователя должен быть выбор из 4 команд:

Добавить товар

Убрать товар

Показать список

Закончить

При добавлении товара пользователь так-же указывает его количество, если пользователь добавляет товар, который уже есть в списке, то его количество должно добавиться к количеству существующего товара

```python
def remove_item(items):
  """
  Удаляет товар из списка покупок
  """
  if not items:
    print("Список товаров пустой")
    return items

  item=input("Введите название для удаления")

  if item in items:
    del items[item]
    print("Удалили товар из списка покупок")
    return items
  else:
    print("Товара нет в списке")
    return items
```

```python
def add_item(items):
  """
  Добавляет товар в список покупок
  """

  item=input("Введите название товара")
  if not item:
    print("Название товара не может быть пустым")
    return items

  quntity=input("Введите количество товара")
  if(quntity.isnumeric()):
    quntity=int(quntity)
    if(quntity<=0):
      print("Введите корректное количество товара")
      return items
    if item in items:
      items[item]+=quntity
      print("Изменили количество товара в списке")
      return items
    else:
      items[item]=quntity
      print("Добавили новый товар")
      return items
  else:
    print("Введите корректное количество товара")
    return items
```

```python
def show_list(items):
  if not items:
    print("Список товаров пустой")
    return items

  for item in items:
    print("Товар:", item, ", количество:",items[item])
```

```python
def shoping_list():
  """
  Дает пользователю интрефейс для работы со списком покупок
  """

  items={}

  while True:
    print("Выберите команду")
    print("1 Добавить товар")
    print("2 Убрать товар")
    print("3 Показать список покупок")
    print("4 Закончить")

    comand=input("Введите номер команды от 1 до 4")
    if comand=="1":
     items=add_item(items)
    elif comand=="2":
      items=remove_item(items)
    elif comand=="3":
      show_list(items)
    elif comand=="4":
       print("Заканечиваем работу программы")
       break
    else:
      print("Неверная команда !")
```
