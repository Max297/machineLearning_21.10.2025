```python
#Берем среднее значение столбца Age и подставляем его в незаполненные ячейки в данном столбце
data["Age"].fillna(int(data["Age"].mean()),inplace=True)
```

```python
#Удаляем оставшиеся пустые ячейки
data.dropna(inplace=True)
```

```python
#Группируем матрицу по значению Pclass
data.groupby("Pclass")["Survived"].mean()
```

Начинаем визуализацию данных из датасета

```python
# 1. Покаызваем выживаемость по полу
#Создаем фигуру и указываем ее размер
plt.figure(figsize=(8, 6))
#Получаем проценты виживания для каждого пола
survived_by_sex = data.groupby('Sex')['Survived'].mean()
#Создаем график, указываем полы на ось x и процент выживания на ось y, указываем цвет столбцов
plt.bar(survived_by_sex.index, survived_by_sex.values, color=['pink', 'lightblue'])
plt.title('Выживаемость по полу', fontsize=14, fontweight='bold')
#Добавляем надписи на график
plt.ylabel('Доля выживших', fontsize=12)
#Добавляем сетку на график
plt.grid(axis='y', alpha=0.3)
#Показываем рисунок
plt.show()
```

```python
# 2. Выживаемость по классу
survived_by_class = data.groupby('Pclass')['Survived'].mean()
plt.bar(survived_by_class.index, survived_by_class.values, color=['gold', 'silver', 'brown'])
plt.title('Выживаемость по классу')
plt.xlabel('Класс')
plt.ylabel('Доля выживших')
plt.ylim(0, 1)
plt.show()
```

```python
# Визуализация выживаемости по полу и классу одновременно
plt.figure(figsize=(10, 6))

# Группируем данные по полу и классу
grouped = data.groupby(['Sex', 'Pclass'])['Survived'].mean().unstack()

# Создаем столбчатую диаграмму
x = np.arange(len(grouped.index))
width = 0.25

for i, pclass in enumerate(grouped.columns):
    plt.bar(x + i * width, grouped[pclass], width, label=f'Класс {pclass}')

plt.xlabel('Пол')
plt.ylabel('Доля выживших')
plt.title('Выживаемость по полу и классу')
plt.xticks(x + width, grouped.index)
plt.legend()
plt.ylim(0, 1)
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Создаем фигуру с несколькими изображенями
# Сначала указываем количество рядов, поток количество столбцов и в конце размер
fig, axes = plt.subplots(1, 2, figsize=(15, 10))
# Общий заголовок
fig.suptitle('Визуализация данных Титаника', fontsize=16, fontweight='bold')

# 1. Распределение возрастов
# Указываем на каком рисунке будет данный график
axes[0].hist(data['Age'], bins=20, alpha=0.7, color='lightgreen', edgecolor='black')
axes[0].set_title('Распределение возрастов')
axes[0].set_xlabel('Возраст')
axes[0].set_ylabel('Количество')

# 2. Выживаемость по возрасту (гистограмма)
survived_age = data[data['Survived'] == 1]['Age']
not_survived_age = data[data['Survived'] == 0]['Age']

#Мы можем поместить 2 параметра на ось x
axes[1].hist([survived_age, not_survived_age], bins=15, alpha=0.7,
                label=['Выжил', 'Не выжил'], color=['green', 'red'])
axes[1].set_title('Распределение возрастов по выживаемости')
axes[1].set_xlabel('Возраст')
axes[1].set_ylabel('Количество')
axes[1].legend()



plt.tight_layout()
plt.show()
```
