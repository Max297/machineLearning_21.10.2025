```python
import random
from datetime import datetime
```

```python
class Message:
  def __init__(self,text,userId,time,reply=-1):
    self.text=text
    self.userId=userId
    self.time=time
    self.id=random.randint(1,10000000000)
    self.replyId=reply
```

```python
class Post:
  def __init__(self,title,text,userId,time):
    self.title=title
    self.text=text
    self.userId=userId
    self.postId=random.randint(1,10000000000)
    self.time=time
    self.messages={}

  def add_message(self,text,userId,time,reply=-1):
    newMessage= Message(text,userId,time,reply)
    self.messages[newMessage.id]=newMessage
```

```python
class Forum:
  def __init__(self):
    self.posts={}
    self.users={}

  def add_user(self,login,password,email):
    sessionId=random.randint(1,10000000000)
    user= AuthUser(sessionId,login,password,email)
    self.users[user.userId]=user

  def add_moderator(self,login,password,email):
    sessionId=random.randint(1,10000000000)
    user= Moderator(sessionId,login,password,email)
    self.users[user.userId]=user

  def add_post(self,userId,title,text):
    date=datetime.now()
    post=Post(title,text,userId,date)
    self.posts[post.postId]=post

  def create_comment(self,userId,postId,text,messageId=-1):
    if (postId in self.posts) and (userId in self.users):
      workPost=self.posts.get(postId)
      workPost.add_message(text,userId,datetime.now(),messageId)

  def delete_comment(self,userId,commentId,postId):
    if (postId in self.posts) and (userId in self.users):
      if (commentId in self.posts[postId].messages) and ("delete_comments" in self.users[userId].permissions) :
        del self.posts[postId].messages[commentId]

  def delete_posts(self,userId,postId):
    print(postId)
    if (postId in self.posts) and (userId in self.users):
      if ("delete_posts" in self.users[userId].permissions) :
        del self.posts[postId]



```

```python
class BasicUser:
  def __init__(self, session_id):
    self.session_id = session_id

  def auth (self,password,login="123123"):
    print("Пытаемся авторизоваться")

  def open_disscussion(self,forumUrl="/test"):
    print("Открываем беседу по url", forumUrl)
```

```python
class AuthUser(BasicUser):
  def __init__(self, session_id,login,password,email):
    self.login=login
    self.password=password
    self.email=email
    self.userId=random.randint(1,100000)
    super().__init__(session_id)
    self.permissions = [
        "create_posts",
        "write_comments",
        "like_posts",
        "write_dm"
    ]

  def view_permissions(self):
    for i in self.permissions:
      print("Пользователь может:", i)

    return self.permissions

  def like_post(self, postId):
    print("Реагируем на пост:", postId)


```

```python
class Moderator(AuthUser):
  def __init__(self, session_id,login,password,email):
    super().__init__(session_id,login,password,email)
    self.permissions = [
        "create_posts",
        "write_comments",
        "like_posts",
        "write_dm",
        "delete_posts",
        "delete_comments"
    ]
```
